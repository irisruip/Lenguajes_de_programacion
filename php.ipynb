```python
'''<?php
// inicializacion de variables

/* php es de tipado debil, esto significa que no es necesario declarar el tipo de dato de una variable.
para inicializar una variable debemos escribir su nombre el cual debe comenzar con el simbolo de dolar,
seguido de un signo de igual y el valor que queremos asignarle a la variable. php le asignara el tipo de dato
a la variable dependiendo del valor que le asignemos. php tambien es de tipado dinamico, esto significa que
podemos cambiar el valor de una variable en cualquier momento y php le asignara el nuevo tipo de dato a la variable.
ahora vamos con algunos ejemplo de inicializacion de variables.
*/

// variables numericas

/*
las variables numericas son aquellas que almacenan numeros enteros o decimales.
*/

$numero = 5;
$numero_decimal = 5.5;

// operaciones aritmeticas

/*
las operaciones aritmeticas son operaciones matematicas que se pueden realizar con variables numericas.
en php contamos con los operadores aritmeticos de suma (+), resta (-), multiplicacion (*), division (/) y modulo (%).
*/

// variables de texto

/*
las variables de texto son aquellas que almacenan cadenas de texto.
*/

$texto = "hola mundo";

// variables booleanas

/*
las variables booleanas son aquellas que almacenan valores de verdadero o falso.
*/

$verdadero = true;
$falso = false;

// parseo de variables

/*
el parseo de variables es la conversion de una variable de un tipo de dato a otro.
php nos permite hacer conversiones de tipo de dato de manera implicita y explicita.
*/

// conversion implicita

/*
la conversion implicita es cuando php convierte una variable de un tipo de dato a otro
automaticamente, esto se da cuando se realiza una operacion aritmetica entre dos variables
de diferente tipo de dato, php convierte la variable de menor jerarquia a la de mayor jerarquia.
*/

$numero = 5;
$numero_decimal = 5.5;

$resultado = $numero + $numero_decimal;
echo "-conversion implicita al sumar 5 con 5.5-";
echo "<br> <br>";
var_dump($resultado);
echo "<br> <br>";

// conversion explicita

/*
la conversion explicita es cuando php convierte una variable de un tipo de dato a otro
de manera manual, esto se da cuando utilizamos una funcion de conversion de tipo de dato
como (int), (float), (string), (bool), etc.
*/

$numero = 5;

$numero = (float) $numero;
echo "-conversion explicita de 5 entero a 5 decimal-";
echo "<br> <br>";
var_dump($numero);
echo "<br> <br>";

// constantes

/*
las constantes son variables que no pueden cambiar su valor una vez que han sido declaradas.
para declarar una constante debemos utilizar la funcion define, la cual recibe dos parametros,
el nombre de la constante y el valor que queremos asignarle a la constante. las constantes
se escriben en mayusculas y no llevan el simbolo de dolar.
*/
echo "-impresion de una constante-";
echo "<br> <br>";
define("CONSTANTE", "soy una constante");

var_dump(CONSTANTE);
echo "<br> <br>";
?>'''
```

```python
'''
<?php
// condicionales

/*
los condicionales son estructuras de control que nos permiten evaluar una o varias condiciones
y ejecutar un bloque de codigo si la condicion se cumple. en php contamos con los condicionales
if, elseif, else, switch y case. para poder usarlos debemos escribir la palabra reservada seguida
de un parentesis y dentro del parentesis la condicion que queremos evaluar, seguido de una llave de
apertura y dentro de las llaves el bloque de codigo que queremos ejecutar si la condicion se cumple.
*/

// if

/*
if es una estructura de control que nos permite evaluar una condicion y ejecutar un bloque de codigo
si la condicion se cumple. if recibe un parametro que es la condicion que queremos evaluar.
*/

$numero = 5;
echo "-impresion de un condicional que evalua si un numero es 5 usando if-";
echo "<br> <br>";
if ($numero == 5) {
    echo "el numero es 5";
}

echo "<br> <br>";

// elseif

/*
elseif es una estructura de control que nos permite evaluar una condicion si la condicion del if
no se cumple. elseif recibe un parametro que es la condicion que queremos evaluar.
*/

$numero = 5;
echo "-impresion de un condicional que evalua si un numero es 4 o 5 usando elseif-";
echo "<br> <br>";
if ($numero == 4) {
    echo "el numero es 4";
} elseif ($numero == 5) {
    echo "el numero es 5";
}

echo "<br> <br>";

// else

/*
else es una estructura de control que nos permite ejecutar un bloque de codigo si la condicion del if
no se cumple. else no recibe parametros.
*/
echo "-impresion de un condicional que evalua si un numero es 4 o 5 usando else-";
echo "<br> <br>";
$numero = 5;

if ($numero == 4) {
    echo "el numero es 4";
} else {
    echo "el numero no es 4";
}

echo "<br> <br>";

// switch y case

/*
switch y case son estructuras de control que nos permiten evaluar una condicion y ejecutar un bloque de codigo
dependiendo del valor de la condicion. switch recibe un parametro que es la condicion que queremos evaluar y
case recibe un parametro que es el valor que queremos comparar con la condicion del switch. su sintaxis es la
siguiente:

switch (condicion) {
    case valor1:
        // bloque de codigo
        break;
    case valor2:
        // bloque de codigo
        break;
    default:
        // bloque de codigo
}

en este caso si la condicion del switch es igual a valor1 se ejecutara el bloque de codigo del case valor1,
si la condicion del switch es igual a valor2 se ejecutara el bloque de codigo del case valor2, si la condicion
del switch no es igual a ninguno de los valores de los case se ejecutara el bloque de codigo del default.
*/

$numero = 5;
echo "-impresion de un condicional que evalua si un numero es 4 o 5 usando switch y case-";
echo "<br> <br>";
switch ($numero) {
    case 4:
        echo "el numero es 4";
        break;
    case 5:
        echo "el numero es 5";
        break;
    default:
        echo "el numero no es 4 ni 5";
}

echo "<br> <br>";

//bloques try y catch

/*
los bloques try y catch son estructuras de control que nos permiten manejar excepciones en php.
try es un bloque de codigo que contiene el codigo que queremos ejecutar y catch es un bloque de codigo
que contiene el codigo que queremos ejecutar si se produce una excepcion. en php las excepciones son
errores que se producen durante la ejecucion del codigo y que pueden ser capturados y manejados por nosotros.
*/

// ejemplo de excepcion

/*
en este caso vamos a intentar dividir un numero por cero, lo cual nos va a producir una excepcion
ya que no se puede dividir un numero por cero. para manejar esta excepcion vamos a utilizar un bloque
try y catch.
*/

echo "-bloque try y catch que maneja una excepcion al dividir un numero por cero-";
echo "<br><br>";

try {
    $numero = 5 / 0;
} catch (Error $e) {
    echo "Se produjo un error: " . $e->getMessage();
}

echo "<br><br>";
?>'''
```

```python
'''
<?php
// bucles

/*
los bucles son estructuras de control que nos permiten ejecutar un bloque de codigo varias veces
dependiendo de una condicion. en php contamos con los bucles while, do while, for y foreach. para
poder usarlos debemos escribir la palabra reservada seguida de un parentesis y dentro del parentesis
la condicion que queremos evaluar, seguido de una llave de apertura y dentro de las llaves el bloque
de codigo que queremos ejecutar varias veces.
*/

// while

/*
while es una estructura de control que nos permite ejecutar un bloque de codigo mientras una condicion
se cumpla. while recibe un parametro que es la condicion que queremos evaluar. su sintaxis es la siguiente:

while (condicion) {
    // bloque de codigo
}

en este caso si la condicion se cumple se ejecutara el bloque de codigo, si la condicion no se cumple
se saldra del bucle.
*/

$numero = 0;
echo "-ciclo while que imprime numeros del 0 al 4-";
echo "<br> <br>";
while ($numero < 5) {
    echo $numero;
    $numero++;
}
// es importante que la condicion se pueda cumplir en algun momento, de lo contrario se creara un bucle infinito.

echo "<br> <br>";

// do while

/*
do while es una estructura de control que nos permite ejecutar un bloque de codigo al menos una vez
y luego mientras una condicion se cumpla. do while no recibe parametros. su sintaxis es la siguiente:

do {
    // bloque de codigo
} while (condicion);

en este caso se ejecutara el bloque de codigo al menos una vez y luego se ejecutara mientras la condicion
se cumpla.
*/
echo "-ciclo do while que imprime numeros del 0 al 4-";
$numero = 0;
echo "<br> <br>";
do {
    echo $numero;
    $numero++;
} while ($numero < 5);

echo "<br> <br>";

// for

/*
for es una estructura de control que nos permite ejecutar un bloque de codigo un numero determinado de veces.
for recibe tres parametros, el primero es la inicializacion de la variable, el segundo es la condicion que queremos
evaluar y el tercero es el incremento o decremento de la variable. su sintaxis es similar a la de java y c y es la siguiente:

for (inicializacion; condicion; incremento) {
    // bloque de codigo
}

en este caso se ejecutara el bloque de codigo un numero determinado de veces dependiendo de la condicion.
*/
echo "-ciclo for que imprime numeros del 0 al 4-";
echo "<br> <br>";
for ($numero = 0; $numero < 5; $numero++) {
    echo $numero;
}

echo "<br> <br>";

// foreach

/*
foreach es una estructura de control que nos permite recorrer un vector y ejecutar un bloque de codigo por cada
elemento del vector. foreach recibe dos parametros, el primero es el vector que queremos recorrer y el segundo es
la variable que queremos utilizar para acceder a los elementos del vector. su sintaxis es la siguiente:

foreach ($vector as $elemento) {
    // bloque de codigo
}

en este caso se ejecutara el bloque de codigo por cada elemento del vector.
*/
echo "-ciclo foreach que imprime los elementos de un vector-";
echo "<br> <br>";
$vector = [1, 2, 3, 4, 5];

foreach ($vector as $elemento) {
    echo $elemento;
}

echo "<br> <br>";

// vectores y matrices'

/*
los vectores y matrices son estructuras de datos que nos permiten almacenar varios valores en una sola variable.
en php los vectores y matrices son arreglos indexados, esto significa que los valores se almacenan en una posicion
especifica y se acceden a traves de un indice. para declarar un vector debemos escribir el nombre de la variable
seguido de un signo de igual y entre corchetes los valores que queremos almacenar en el vector. para declarar una
matriz debemos escribir el nombre de la variable seguido de un signo de igual y entre corchetes los valores que queremos
almacenar en la matriz, ademas de un segundo indice que nos permitira acceder a los valores de la matriz. en php los
indices de los arreglos comienzan en 0.
*/

// vector
echo "-declaracion de un vector y acceso a un elemento-";
echo "<br> <br>";
$vector = [1, 2, 3, 4, 5];

echo $vector[3];
echo "<br> <br>";

// matriz
echo "-declaracion de una matriz y acceso a un elemento-";
echo "<br> <br>";
$matriz = [[1, 2], [3, 4], [5, 6]];

echo $matriz[1][0];
echo "<br> <br>";

/* otra forma de declarar un vector o matriz es utilizando la funcion array, la cual recibe como parametros
los valores que queremos almacenar en el vector o matriz. */

// vector
echo "-declaracion de un vector usando la funcion array y acceso a un elemento-";
echo "<br> <br>";
$vector = array(1, 2, 3, 4, 5);

echo $vector[2];

echo "<br> <br>";

// matriz
echo "-declaracion de una matriz usando la funcion array y acceso a un elemento-";
echo "<br> <br>";
$matriz = array(array(1, 2), array(3, 4), array(5, 6));

echo $matriz[0][1];

echo "<br> <br>";

//vectores y matrices asociativos

/*
los vectores y matrices asociativos son estructuras de datos que nos permiten almacenar varios valores en una sola variable
pero a diferencia de los arreglos indexados, en los arreglos asociativos los valores se almacenan en una clave y se acceden
a traves de la clave (parecido a los diccionarios en python). para declarar un vector asociativo debemos escribir el nombre
de la variable seguido de un signo de igual y entre corchetes la clave y el valor que queremos almacenar en el vector asociativo.
para declarar una matriz asociativa debemos escribir el nombre de la variable seguido de un signo de igual y entre corchetes la
clave y el valor que queremos almacenar en la matriz asociativa, ademas de un segundo indice que nos permitira acceder a los valores
de la matriz asociativa.
*/

// vector asociativo
echo "-declaracion de un vector asociativo y acceso a un elemento-";
echo "<br> <br>";
$vector_asociativo = ["clave1" => 1, "clave2" => 2, "clave3" => 3];

echo $vector_asociativo["clave1"];

echo "<br> <br>";

// matriz asociativa
echo "-declaracion de una matriz asociativa y acceso a un elemento-";
echo "<br> <br>";
$matriz_asociativa = ["fila1" => ["columna1" => 1, "columna2" => 2], "fila2" => ["columna1" => 3, "columna2" => 4], "fila3" => ["columna1" => 5, "columna2" => 6]];

echo $matriz_asociativa["fila3"]["columna1"];

echo "<br> <br>";

// recorrido de vectores y matrices

/*
aqui usaremos lo ya aprendido en bucles para recorrer vectores y matrices. en el caso de los vectores
usaremos un bucle for para recorrer el vector y en el caso de las matrices usaremos dos bucles for anidados
para recorrer la matriz.
*/

// recorrido de vector
echo "-recorrido de un vector-";
echo "<br> <br>";
$vector = [1, 2, 3, 4, 5];

for ($i = 0; $i < count($vector); $i++) {
    echo $vector[$i];
}

echo "<br> <br>";

// recorrido de matriz

$matriz = [[1, 2], [3, 4], [5, 6]];
echo "-recorrido de una matriz-";
echo "<br> <br>";
for ($i = 0; $i < count($matriz); $i++) {
    for ($j = 0; $j < count($matriz[$i]); $j++) {
        echo $matriz[$i][$j];
    }
}

echo "<br> <br>";

/* hay varias maneras de recorrer matrices como los bucles foreach anidados, pero en este caso usamos dos bucles for 
anidados para recorrer la matriz. */
?>'''
```
